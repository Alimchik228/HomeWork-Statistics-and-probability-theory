#### 4. Рост взрослого населения города X имеет нормальное распределение.
#### Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
#### Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
#### а). больше 182 см
#### б). больше 190 см
#### в). от 166 см до 190 см
#### г). от 166 см до 182 см
#### д). от 158 см до 190 см
#### е). не выше 150 см или не ниже 190 см
#### ё). не выше 150 см или не ниже 198 см
#### ж). ниже 166 см.

```python
# a) 
a = (182 - 174)/8
print(a)
print('Вероятность А: ', (1 - 0.8413))
# б) 
a = (190 - 174)/8
print(a)
print('Вероятность Б: ', (1 - 0.97725))
# в) 
a = (166 - 174)/8
print(a)
print('Вероятность В: ', (0.9772 - 0.1587))
# г) 
print('Вероятность Г: ', (0.8413 - 0.1587) )
# д) 
a = (158 - 174)/8
print(a)
print('Вероятность Д: ', (0.97725 - (1 - 0.97725)))
# е) 
a = (150 - 174)/8
print(a)
print('Вероятность Е: ', (1 - 0.99865) + (1 - 0.97725))
# ё) 
a = (198 - 174)/8
print(a)
print('Вероятность Ё: ', (1 - 0.99865) + (1 - 0.99865))
# ж) 
a = (166 - 174)/8
print(a)
print('Вероятность Ж: ', 0.1587)
```
